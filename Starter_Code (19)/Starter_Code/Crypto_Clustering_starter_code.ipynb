```python
# Import required libraries and dependencies
import pandas as pd
import matplotlib.pyplot
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaled_data = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
scaled_df_normalized = pd.DataFrame(scaled_data, columns=df_market_data.columns)
scaled_df_normalized['coin_id'] = df_market_data.index
scaled_df_normalized.set_index('coin_id', inplace=True)


# Copy the crypto names from the original data

# Set the coinid column as index



# Display sample data
scaled_df_normalized.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values_list = list(range(1, 12))
k_values_list
```

```python
# Create an empty list to store the inertia values


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

inertia_values = []

for k in k_values_list:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(scaled_df_normalized)
    inertia_values.append(model.inertia_)

inertia_values
```

```python
# Create a dictionary with the data to plot the Elbow curve


# Create a DataFrame with the data to plot the Elbow curve
# Creating a dictionary with the data for plotting the Elbow curve
elbow_data = {
    "k": k_values_list,
    "inertia": inertia_values
}

# Creating a DataFrame with the elbow data
elbow_df = pd.DataFrame(elbow_data)

elbow_df
```

```python
# # Plot a line chart with all the inertia values computed with 
# # the different values of k to visually identify the optimal value for k.
# # Plotting the elbow curve using the inertia values from the elbow_df
# plt.figure(figsize=(10, 6))
# plt.plot(elbow_df['k'], elbow_df['inertia'], marker='o', linestyle='--', color='b')
# plt.title('Elbow Curve')
# plt.xlabel('Number of Clusters (k)')
# plt.ylabel('Inertia')
# plt.grid(True)
# plt.show()
elbow_plot = elbow_df.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k_values_list)
elbow_plot
```

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
# Initializing the K-means model with the best value for k (k=4)
best_k = 4
kmeans_best_k = KMeans(n_clusters=best_k, random_state=42)

kmeans_best_k
```

```python
# Fit the K-Means model using the scaled data
kmeans_best_k.fit(scaled_df_normalized)
```

```python
# Predicting the clusters using the scaled data
predicted_clusters = kmeans_best_k.predict(scaled_df_normalized)

# Printing the resulting array of cluster values
predicted_clusters
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)

pca_model
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.

# View the first five rows of the DataFrame. 
# Using the PCA model to reduce the data to three principal components
crypto_pca_data = pca_model.fit_transform(scaled_df_normalized)

# Creating a DataFrame with the PCA data
pca_df = pd.DataFrame(data=crypto_pca_data, columns=['PC1', 'PC2', 'PC3'], index=crypto_df_copy.index)

# Displaying the first five rows of the PCA DataFrame
pca_df.head()

```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.

pca_model.explained_variance_ratio_
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values_pca_list = list(range(1, 12))
k_values_pca_list
```

```python
# Create an empty list to store the inertia values
inertia_values_pca = []

inertia_values_pca


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_values_pca_list:
    model_pca = KMeans(n_clusters=k, random_state=42)
    model_pca.fit(pca_data_with_names)
    inertia_values_pca.append(model_pca.inertia_)

inertia_values_pca
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {
    "k": k_values_pca_list,
    "inertia": inertia_values_pca
}

elbow_data_pca

# Create a DataFrame with the data to plot the Elbow curve
elbow_pca_df = pd.DataFrame(elbow_data_pca)

elbow_pca_df
```

```python
# # Plot a line chart with all the inertia values computed with 
# # the different values of k to visually identify the optimal value for k.
# # Plotting the elbow curve for PCA data using the inertia values from the elbow_pca_df
# plt.figure(figsize=(10, 6))
# plt.plot(elbow_pca_df['k'], elbow_pca_df['inertia'], marker='o', linestyle='--', color='b')
# plt.title('Elbow Curve for PCA Data')
# plt.xlabel('Number of Clusters (k)')
# plt.ylabel('Inertia')
# plt.grid(True)
# plt.show()

elbow_plot_pca = elbow_pca_df.hvplot.line(x="k", y="inertia", title="Elbow Curve Using PCA Data", xticks=k_values_list)
elbow_plot_pca
```
